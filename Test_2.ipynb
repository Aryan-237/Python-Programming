```python
class Student:

    school_name = "UPES"

    def __init__(self ,name ,age):
        self.name = name
        self.age = age
   

c1= Student("Aryan" , 18)
c2= Student("Zaid" , 19)
print("School:" , c1.school_name)
print("Student 1:" , c1.name , c1.age)
print("Student 2:" , c2.name , c2.age)
```

```python
class MathHelper:
    @staticmethod
    def is_even(num):
        return num%2 == 0
    
print(MathHelper.is_even(8))
print(MathHelper.is_even(11))
```
