```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.figure()

plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.bar(x,y)

plt.subplot(2,2,3)
plt.scatter(x,y)

plt.subplot(2,2,4)
plt.pie(y, labels=x)

plt.show()
```
