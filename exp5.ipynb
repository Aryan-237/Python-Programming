```python
text = input("Enter a string: ")

count = 0

for ch in text:
    if 'A' <= ch and  ch <= 'Z':
        count = count + 1

print("Number of capital letters:", count)
```

```python
text = input("Enter a string: ")

count = 0

for ch in text:
    if ch in "AEIOUaeiou":
        count = count + 1

print("Total number of vowels:", count)
```

```python
string = input("Enter the string: ")
sub = input("Enter the substring: ")

count = 0
length = len(sub)

for i in range(len(string) - length + 1):
    if string[i:i+length] == sub:
        count = count + 1

print("Number of occurrences:", count)
```

```python
string = input("Enter the string: ")

string = string.upper()   

count_dict = {}

for ch in string:
    if 'A' <= ch <= 'Z':         
        if ch in count_dict:
            count_dict[ch] += 1
        else:
            count_dict[ch] = 1

for key in sorted(count_dict):
    print(count_dict[key], key)
```

```python
sentence = input("Enter a sentence: ")

sentence = sentence.lower()
words = sentence.split()
unique_words = set(words)

print("Number of unique words:", len(unique_words))
```

```python
n = int(input("Enter number of fruits in each set: "))

s1 = set()
s2 = set()

print("Enter fruits for set s1:")
for i in range(n):
    fruit = input()
    s1.add(fruit)

print("Enter fruits for set s2:")
for i in range(n):
    fruit = input()
    s2.add(fruit)

common_fruits = s1 & s2
only_s1 = s1 - s2
all_fruits = s1 | s2

print("Fruits in both s1 and s2:", common_fruits)
print("Fruits only in s1:", only_s1)
print("Count of all fruits:", len(all_fruits))
```

```python
S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Set S1:", S1)
print("Set S2:", S2)

print("Union:", S1 | S2)
print("Intersection:", S1 & S2)
print("Difference (S1 - S2):", S1 - S2)
print("Difference (S2 - S1):", S2 - S1)
print("Symmetric Difference:", S1 ^ S2)
```
