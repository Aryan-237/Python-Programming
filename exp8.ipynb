```python
with open("name.txt","a") as f:
    pass

with open("name.txt","r") as f:
    names=[i.strip() for i in f if i.strip()]

if names:
    print(len(names))
    print(sum(1 for i in names if i[0].lower() in "aeiou"))
    print(max(names,key=len))
else:
    print("File is empty")
```

```python
with open("numbers.txt","a") as f:
    pass

with open("numbers.txt","r") as f:
    nums=[int(i.strip()) for i in f if i.strip().isdigit()]

if nums:
    print(max(nums))
    print(sum(nums)/len(nums))
    print(sum(1 for i in nums if i>100))
else:
    print("No valid numbers")
```

```python
with open("city.txt","a") as f:
    pass

with open("city.txt","r") as f:
    cities=[i.split() for i in f if i.strip()]

if cities:
    for i in cities:
        print(i)

    for i in cities:
        if float(i[1])>10:
            print(i[0])

    print(sum(float(i[2]) for i in cities))
else:
    print("File is empty")
```

```python
n=int(input())
for _ in range(n):
    try:
        a,b=input().split()
        print(int(a)//int(b))
    except Exception as e:
        print("Error Code:",e)
```

```python
class EmptyFileError(Exception):
    pass

try:
    with open("data.txt","a") as f:
        pass

    with open("data.txt","r") as f:
        d=f.read().strip()
        if not d:
            raise EmptyFileError("File is empty")
        print(d)
except Exception as e:
    print(e)
```
