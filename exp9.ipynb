```python
class Student:
    def __init__(self, name, sap, phy, chem, math):
        self.name = name
        self.sap = sap
        self.phy = phy
        self.chem = chem
        self.math = math

    def display(self):
        print(self.name, self.sap, self.phy, self.chem, self.math)

    def percentage(self):
        return (self.phy + self.chem + self.math) / 3

    def result(self):
        if self.phy > 40 and self.chem > 40 and self.math > 40:
            print("Pass")
        else:
            print("Fail")

def class_avg(students):
    print(sum(s.percentage() for s in students)/len(students))

s1 = Student("Aman",101,80,70,90)
s2 = Student("Riya",102,30,75,85)
s3 = Student("Aryan",103,50,55,65)

students = [s1,s2,s3]

for s in students:
    s.display()
    print(s.percentage())
    s.result()

class_avg(students)
```

```python
class A:
    def showA(self):
        print("Class A")

class B(A):
    def showB(self):
        print("Class B")

class C(B):
    def showC(self):
        print("Class C")

obj = C()
obj.showA()
obj.showB()
obj.showC()
```

```python
class Parent:
    def show(self):
        print("Parent class")

class Child(Parent):
    def show(self):
        print("Child class")

obj = Child()
obj.show()
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def display(self):
        print(self.x, self.y)

p1 = Point(10,20)
p2 = Point(12,15)

p3 = p1 + p2
p3.display()
```
